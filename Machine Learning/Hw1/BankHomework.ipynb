```python
import csv 
import numpy
import pandas as pd 
from sklearn.preprocessing import LabelEncoder

bank = pd.read_csv("bank.csv",delimiter=";",header='infer')

LB  = LabelEncoder()

def GetMapplist(lb,df,para) :  #傳入encoder,dataframe,para(你要的欄位字串)
    lb.fit_transform(df[para])
    lb_name_mapping_month = dict(zip(lb.classes_, lb.transform(lb.classes_)))
    print(lb_name_mapping_month)


GetMapplist(LB,bank,'job')
#{'admin.': 0, 'blue-collar': 1, 'entrepreneur': 2, 'housemaid': 3, 'management': 4, 'retired': 5, 
#'self-employed': 6, 'services': 7, 'student': 8, 'technician': 9, 'unemployed': 10, 'unknown': 11}

GetMapplist(LB,bank,'marital')
#{'divorced': 0, 'married': 1, 'single': 2}

GetMapplist(LB,bank,'education')
#{'primary': 0, 'secondary': 1, 'tertiary': 2, 'unknown': 3}

GetMapplist(LB,bank,'default')
#{'no': 0, 'yes': 1}

GetMapplist(LB,bank,'housing')
#{'no': 0, 'yes': 1}

GetMapplist(LB,bank,'loan')
#{'no': 0, 'yes': 1}

GetMapplist(LB,bank,'contact')
#{'cellular': 0, 'telephone': 1, 'unknown': 2}

GetMapplist(LB,bank,'month')
#{'apr': 0, 'aug': 1, 'dec': 2, 'feb': 3, 'jan': 4, 'jul': 5, 'jun': 6, 'mar': 7, 'may': 8, 'nov': 9, 'oct': 10, 'sep': 11}

GetMapplist(LB,bank,'poutcome')
#{'failure': 0, 'other': 1, 'success': 2, 'unknown': 3}

GetMapplist(LB,bank,'y')
#{'no': 0, 'yes': 1}

bank['job'] = LB.fit_transform(bank['job'])
bank['marital'] = LB.fit_transform(bank['marital'])
bank['education'] = LB.fit_transform(bank['education'])
bank['default'] = LB.fit_transform(bank['default'])
bank['housing'] = LB.fit_transform(bank['housing'])
bank['loan'] = LB.fit_transform(bank['loan'])
bank['contact'] = LB.fit_transform(bank['contact'])
bank['month'] = LB.fit_transform(bank['month'])
bank['poutcome'] = LB.fit_transform(bank['poutcome'])
bank['y'] = LB.fit_transform(bank['y'])

#solution2
#bank['y'] = bank['y'].replace({'no':0,'yes':1})


bank.head(100)
```
