```python
import csv 
import numpy
import pandas as pd 
from sklearn.preprocessing import LabelEncoder
import numpy as np
np.set_printoptions(threshold=np.inf)
#讀檔
BigMart = pd.read_csv("Train.csv",delimiter=",",header='infer')
#先查看那些欄位需要做replace
# print(len(BigMart['Item_Identifier'].unique())) 
# print(BigMart['Item_Weight'].unique())
# print(BigMart['Item_Visibility'].unique())
# print(BigMart['Item_Type'].unique())
# print(BigMart['Item_MRP'].unique())
# print(BigMart['Outlet_Identifier'].unique())
# print(BigMart['Outlet_Establishment_Year'].unique())
# print(BigMart['Outlet_Location_Type'].unique())
# print(BigMart['Outlet_Type'].unique())
# print(BigMart['Item_Outlet_Sales'].unique())

#Item_Fat_Content 欄位中相同意思的資料進行整合
BigMart['Item_Fat_Content'] = BigMart['Item_Fat_Content'].replace({'low fat':'Low Fat','LF':'Low Fat','low fat':'Low Fat','reg':'Regular'})

def GetMapplist(lb,df,para) :  #傳入encoder,dataframe,para(你要的欄位字串)
    lb.fit_transform(df[para])
    df[para] = lb.fit_transform(df[para])  #編碼完成後存入原本的dataframe
    lb_name_mapping = dict(zip(lb.classes_, lb.transform(lb.classes_)))
    print(lb_name_mapping)
    
LB = LabelEncoder()

#下面這行能將資料進行取代  EX out049>>049   可將out擷取掉   
BigMart['Outlet_Identifier'] = BigMart['Outlet_Identifier'].str.replace('OUT','')
GetMapplist(LB,BigMart,'Outlet_Identifier')
#BigMart['Outlet_Identifier'] {'010': 0, '013': 1, '017': 2, '018': 3, '019': 4, '027': 5, 
#       '035': 6, '045': 7, '046': 8, '049': 9}

GetMapplist(LB,BigMart,'Item_Fat_Content')
#BigMart['Item_Fat_Content']  {'Low Fat': 0, 'Regular': 1}

GetMapplist(LB,BigMart,'Item_Type')
#BigMart['Item_Type']
#{'Baking Goods': 0, 'Breads': 1, 'Breakfast': 2, 'Canned': 3, 'Dairy': 4, 'Frozen Foods': 5, 'Fruits and Vegetables': 6
# ,'Hard Drinks': 7, 'Health and Hygiene': 8, 'Household': 9, 'Meat': 10, 'Others': 11, 'Seafood': 12, 'Snack Foods': 13, 
# 'Soft Drinks': 14, 'Starchy Foods': 15}

#將空值進行填充
BigMart = BigMart.fillna({'Outlet_Size':0})
#編碼用replace方式
BigMart['Outlet_Size'] = BigMart['Outlet_Size'].replace({'Small':1,'Medium':2,'High':3})
#BigMart['Outlet_Size']  {NaN:0,'Small':1,'Medium':2,'High':3}


GetMapplist(LB,BigMart,'Outlet_Location_Type')
#BigMart['Outlet_Location_Type']  {'Tier 1': 0, 'Tier 2': 1, 'Tier 3': 2}


GetMapplist(LB,BigMart,'Outlet_Type')
#BigMart['Outlet_Type'] {'Grocery Store': 0, 'Supermarket Type1': 1, 'Supermarket Type2': 2, 'Supermarket Type3': 3}


#以下code用於coulumn欄位儲存  可供未來使用
# bcolumn =[]
# for i in BigMart.columns:
#     bcolumn.append(i)
# print(bcolumn)
    
BigMart.head(20)
```
